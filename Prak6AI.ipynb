```python
#memanggil library
import pandas as pd
import numpy as np

#membaca data kelulusan mahasiswa
mhs = pd.read_excel('DATA_LAYANAN_NETWORK.xlsx')
mhs.head()
```

```python
#membuat atribut reguler
x = mhs.drop(['LS'],axis = 1)
x.head()
```

```python
#membuat atribut target/class
y = mhs['LS']
y.head(10)
```

```python
#import gausian naive bayes model
from sklearn.naive_bayes import GaussianNB
```

```python
#memanggil fungsi gausian naive bayes
nbc = GaussianNB()
```

```python
#memasukkan data training ke dalam fungsi klasifikasi naive bayes
data_training = nbc.fit(x,y)
```

```python
y_predict = data_training.predict(x)
print(y_predict)
```

```python
from re import I
#melakukan prediksi pada data input
SK=1
MS=1
LI=0
JK=1

Data_testing1=[[SK,MS,LI,JK]]
Data_testing1
```

```python
y_pred= data_training.predict(Data_testing1)
print(y_pred)
```

```python
#mencetak hasil
if y_pred ==0:
  hasil ='Tidak'
elif y_pred == 1:
  hasil ='iya'
else :
  hasil = 'Error'
print('Hasil Prediksi: ',hasil)
```

```python
from sklearn.metrics import accuracy_score
print('Nilai akurasi = %0.2f' % accuracy_score(y,y_predict))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y,y_predict))
```
